```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Histogram
#Scatter plot
#Area chart
#Heatmap

                                    #HISTOGRAM
df = pd.read_csv('seattle-weather.csv').head(20)
df
```

```python
date = df['date']
precipitation = df['precipitation']

#Histogram showing precipitation
plt.hist(precipitation, color='yellow', edgecolor='black')

plt.title(' Histogram representing precitation')
plt.show()
```

```python
#Scatter plot of min and max temp
min_temp = df['temp_min']
max_temp = df['temp_max']
plt.figure(figsize=(15, 7))
plt.scatter(date, max_temp, label = "Max_Temp")
plt.scatter(date, min_temp, label="Min_Temp")


plt.title('Scatter plot of minimum and maximum temperatures')
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.ylabel('Temp')
plt.legend()

plt.show()
```

```python
#Area Chart of the wind and precipitation
wind = df['wind']

plt.figure(figsize=(10, 5))
plt.fill_between(date, wind, color='green', label='wind')
plt.fill_between(date, precipitation, color='yellow', label= "Precipitation")


plt.title('Area chart showing wing and precipitation')
plt.xlabel('Date')
plt.ylabel("Precipiatation and wind levels")
plt.xticks(rotation=45)

plt.legend()
plt.show()
```

```python
#HEATMAP
#Heatmap of the min and max temp
temp_data = np.array([min_temp, max_temp])


plt.figure(figsize=(10, 6))
plt.imshow(temp_data, cmap='coolwarm', aspect='auto', interpolation='nearest')

np.arange(len(temp_data))

plt.colorbar(label='Temperature')
plt.title('Heatmap of Min and Max Temperatures')
plt.ylabel('Temperature levels')


plt.show()
```
