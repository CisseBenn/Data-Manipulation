```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pl-table-xg.csv")
df
```

```python
# Below is a line graph of xg against xpoints

sns.lineplot(x="xg", y="xPoints", data=df)
plt.show()
```

```python
# Below is a displot of the teams, xg, and the number of points of the first 5 teams
dp = df.head(3)
df.sort_values(by="pts", ascending=False)
sns.set_theme(style="darkgrid")
sns.displot(data= dp, x="pts", col="xg", row="name", binwidth=3, height=3, facet_kws=dict(margin_titles=True),)
plt.title("Displot of the teams xg")
plt.ylabel("Teams")
plt.xlabel("Points")
plt.show()
```

```python
#Below is a pyplot of the pl table
df.groupby(['name'],).sum().plot(kind="pie", y="pts",autopct='%1.0f%%', legend= False)
```
